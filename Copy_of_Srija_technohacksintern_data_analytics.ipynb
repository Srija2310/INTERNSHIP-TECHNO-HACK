# TASK 3

VISUALIZATION USING **HISTOGRAM**

- Here's a concise explanation of how to create a histogram using Python:

````python
from google.colab import files
uploaded = files.upload()
````

Import Libraries : *Begin by importing necessary libraries. Typically, you'll use matplotlib.pyplot for plotting and pandas for data manipulation*

````python
import pandas as pd
import io
import matplotlib.pyplot as plt
````

*If your data is in a file (e.g., CSV), load it into a pandas DataFrame*

````python
data = pd.read_csv(io.BytesIO(uploaded['train.csv']))
data
````

````python
data.dtypes
````

1.*Plotting the Histogram: Use plt.hist() function from matplotlib.pyplot to create the histogram. Customize the plot as needed by adjusting parameters like number of bins, color, transparency, etc*

2.*Add Labels and Title: Enhance the plot by adding labels to the x-axis, y-axis, and a title*

3.*Display the Plot: Finally, display the histogram using plt.show().*



````python
plt.hist(data['PassengerId'], bins=10, edgecolor='g', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Histogram of Age vs Fare')
plt.show()
````

````python
plt.hist(data['PassengerId'], bins=20, edgecolor='k', alpha=0.7)
plt.xlabel('SibSp')
plt.ylabel('survived')
plt.title('Histogram of SibSp vs Survived')
plt.show()
````
